```python
#Plotting the graph, A country is virtually equal to a function

def floatParser(arr=[]):
    for i,e in enumerate(arr):
        try:
            arr[i] = float(e)
        except(ValueError):            
            arr[i] = -1
    return arr   

#set the x_axis to the years
x_axis = obese_distribution.columns.tolist()[2:]
heading = 'Blood Pressure Prevalence in most Obese countries({}-{})'.format(x_axis[0], x_axis[len(x_axis) - 1])
#configure matplotlib ploting object
plt.figure(figsize=(20,7))
plt.title(heading, fontsize=16, fontweight='bold')
plt.xlabel("Years")
plt.ylabel("Most_obese_countrys")

countryNames = []

#Iterate through each entry in the table rows
for index, row in enumerate(obese_distribution.values.tolist()):
    y_axis = {'countryName': row[1], 'values': floatParser(row[2:])}
    countryNames.append(y_axis['countryName'])
    plt.plot(x_axis, y_axis['values'])    
# plt.legend(countryNames,bbox_to_anchor=(0., 1.02, 1., .102), loc=3,ncol=1, mode="expand", borderaxespad=0.)
plt.legend(countryNames, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

```
